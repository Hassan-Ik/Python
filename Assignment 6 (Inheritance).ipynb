```python
class Circle:
    
    def area(self,radius):
        Area = 3.14*radius**2
        print("Area of Circle is",Area)
        
    def perimeter(self,radius):
        Perimeter = 2*3.14 * radius
        print("Perimeter of Circle is",Perimeter)

red = Circle()

red.area(2)
red.perimeter(3)
```
